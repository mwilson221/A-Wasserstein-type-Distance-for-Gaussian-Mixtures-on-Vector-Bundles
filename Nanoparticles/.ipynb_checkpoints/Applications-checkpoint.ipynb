```python
from moving_frame import *
from geomstats.geometry.hypersphere import Hypersphere
from IPython.display import clear_output

N = 82
n = 2

B = init_fourier_basis(n,N)

manifold = Hypersphere(dim=n*N)

frames=[]

for t in range(1): 
    clear_output(wait=True)
    print(t)
    #1) Load Data
    beta,mode_idx,idx = load_KMKM_data(t)
    
    plot_frame(beta,mode_idx,idx)
    plt.show()

    #2) Align
    aligned_frame = align_by_cluster(beta, idx, N=N)
    
    #3) Project          
    means = [function_project(B,component.q_mean) for component in aligned_frame]
    data = [function_project(B,component.qn) for component in aligned_frame]
        
    #4) Moving Frame
    wgmm  = WrappedGaussianMixture(manifold=manifold)
    wgmm.fit_frame(means,data)
    frames.append(wgmm)
        
np.save("data_files//frames_new_TEST.npy",frames)
```

```python
def energy_perm_test(MW_temp,t,num_perms=100):
    
    N = MW_temp.shape[0]
    
    E_true = (2/(t*(N-t)))*np.sum(MW_temp[:t,t:])-np.sum(MW_temp[:t,:t])/((t)**2)-np.sum(MW_temp[t:,t:])/((N-t)**2) 
    E=np.zeros(num_perms)
    for i in range(num_perms):
        perms=np.random.choice(N,N,replace=False)
        MW_temp=MW_temp[perms,:][:,perms]
        E[i] = (2/(t*(N-t)))*np.sum(MW_temp[:t,t:])-np.sum(MW_temp[:t,:t])/((t)**2)-np.sum(MW_temp[t:,t:])/((N-t)**2) 
        
    # print(E_true)
    # print(E)
    return np.mean(E>E_true),(((t*(N-t)))/((t+(N-t))))*E_true
    
```
